```python
# Import data
filename = 'OMA-project-data.csv'
df = pd.read_csv(filename, dtype={'IMG_': str})

#add decade col
def create_decade(x):
    d = str(x['YEAR'])
    e = d[:3] + "0s"
    return e

df['DECADE'] = df.apply(create_decade, axis=1)

#Update image column with paths to B/W photos
bw_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Black and White Images\\"

def create_bw_path(x):
    return bw_path + str(x['IMG_']) + ".jpg"
    
df['@IMG'] = df.apply(create_bw_path, axis=1)

#Update image column with paths to Typology backgrounds
type_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Typology Backgrounds\\"

def create_type_path(x):
    return type_path + str(x['TYPOLOGY_01']) + ".png"
    
df['@TYPE_BG'] = df.apply(create_type_path, axis=1)

#Drop
df = df.drop(["LINK", "TYPOLOGY_02", "URL STRING", "Unnamed: 11", "IMG_", "@Photos"], axis=1)

# Organise columns
df["TEAM LIST"] = np.nan

cols = df.columns.tolist()
cols = cols[-4:] + cols[:-4]
df = df[cols]

df.head(2)
```

```python
#Create team list
team_cols = df.loc[:, 'TEAM_01':].columns.tolist()
team_cols_len = len(team_cols)

def create_team(x):
    y = []
    for i in team_cols:
        y.append(x[i])
    return y

df["TEAM LIST"] = df.apply(create_team, axis=1)

df.head(2)
```

```python
team_list = df["TEAM LIST"].tolist()

def find_names(name):
    j = []

    for i in team_list:
        j.append(name in i)
    return j

def name_frame(name):
    p = find_names(name)
    q = df[p]
    return q
```

```python
#Set up for paths to Gender backgrounds
gender_img_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Gender Backgrounds\\"

def create_gender_path(o):
    n = gender_img_path + str(o['GENDER']) + ".png"
    return n
```

```python
# Load primary name list
a = ""

a = 'project-people.csv'
primary_df = pd.read_csv(a)

primary_list = primary_df["NAME"].tolist()
print primary_list

primary_df["@Gender"] = primary_df.apply(create_gender_path, axis=1)

save_primary_path = "total_output/primary-people_names.csv"
primary_df.to_csv(save_primary_path, index=False, encoding='utf-8')
print "Saved: " + save_primary_path

primary_df.head()
```

```python
# # Load secondary name list
a = ""

a = 'generations-people.csv'
secondary_df = pd.read_csv(a)

secondary_df["@Gender"] = secondary_df.apply(create_gender_path, axis=1)

secondary_list = secondary_df["NAME"].tolist()
secondary_genders_img = secondary_df["@Gender"].tolist()
print secondary_list
print secondary_genders_img
secondary_df.head()
```

```python
# Process primary list
def process_names(d):
    f = name_frame(d)
    f = f.drop(f.loc[:, 'TEAM_01':], axis=1)
    h = len(f["YEAR"].tolist())
    f["BRANCH KEY"] = d
    f["KEY TOTAL"] = h
    f.sort_values("YEAR")
    print "Done: " + d
    return f

index = []
columns = []
total_df = pd.DataFrame(index=index, columns=columns)

for g in primary_list:
    total_df = total_df.append(process_names(g))  

total_df.head()
```

```python
# Get Secondary key team
def process_secondary_list(r):
    secondary_team = []

    for name in secondary_list:
        if name in r["TEAM LIST"]:
                secondary_team.append(name)
        else:
            pass
    return secondary_team

def process_secondary_list_genders(r):
    secondary_team_genders_paths = []
    length = len(secondary_list)
    for i in range(0, length):
        if secondary_list[i] in r["TEAM LIST"]:
                secondary_team_genders_paths.append(secondary_genders_img[i])
        else:
            pass
    return secondary_team_genders_paths


total_df["SECONDARY TEAM"] = total_df.apply(process_secondary_list, axis=1)
total_df["SECONDARY TEAM GENDER PATHS"] = total_df.apply(process_secondary_list_genders, axis=1)

total_df.head(2)
```

```python
def apply_count(g):
    u = len(g["SECONDARY TEAM"])
    return u

total_df["SECONDARY COUNT"] = total_df.apply(apply_count, axis=1)

total_df.head(2)
```
