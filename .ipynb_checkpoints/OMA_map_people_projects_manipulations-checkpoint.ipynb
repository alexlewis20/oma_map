```python
# Import data
filename = 'OMA-project-data.csv'
df = pd.read_csv(filename, dtype={'IMG_': str})

#add decade col
def create_decade(x):
    d = str(x['YEAR'])
    e = d[:3] + "0s"
    return e

df['DECADE'] = df.apply(create_decade, axis=1)

#Update image column with paths to B/W photos
bw_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Black and White Images\\"

def create_bw_path(x):
    return bw_path + str(x['IMG_']) + ".jpg"
    
df['@IMG'] = df.apply(create_bw_path, axis=1)

#Update image column with paths to Typology backgrounds
type_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Typology Backgrounds\\"

def create_type_path(x):
    return type_path + str(x['TYPOLOGY_01']) + ".png"
    
df['@TYPE_BG'] = df.apply(create_type_path, axis=1)


df = df.drop(["LINK", "TYPOLOGY_02", "URL STRING", "Unnamed: 11", "IMG_", "@Photos"], axis=1)
df.head(2)
```

```python
status_dict = {
               "Commissioned Study": 1,
               "Competition": 2,
               "Completed": 4,
               "Construction": 2,
               "Demolished": 2,
               "Design Development": 2,
               "Lecture": 0,
               "Publication": 4,
               "Study": 0
               }

typology_dict = {
                 "N/A": 0,
                 "Arena": 0,
                 "Branding": 1,
                 "Education": 2,
                 "Exhibition": 2,
                 "Hotel": 2,
                 "Industrial": 0,
                 "Infrastructure": 0,
                 "Landscape": 2,
                 "Library": 3,
                 "Masterplan": 3,
                 "Mixed use": 4,
                 "Museum / Gallery": 3,
                 "Office": 0,
                 "Product Design": 0,
                 "Research": 0,
                 "Residential": 1,
                 "Restaurant / Bar": 0,
                 "Retail": 2,
                 "Scenography": 3,
                 "Theatre": 3,
                 "Tower": 0
                 }

decades_dict = {
                "1970s": 5,
                "1980s": 3,
                "1990s": 1,
                "2000s": 0
                }


dict_dict = {
            "STATUS": status_dict,
            "TYPOLOGY_01": typology_dict,
            "DECADE": decades_dict
            }
```

```python
def set_value (dictionary, key):
    max_dict = max([i for i in dictionary.values()])
    x = dictionary[key] / max_dict
    return x

def importance_factor(y):
    z = []
    
    for k in dict_dict:
        try:
            m = dict_dict[k]
            n = y[k]
            z.append(set_value(m, n))
        except:
            pass
    
    a = float(len(z))
    b = float(sum(z))
    return (b/a)
```

```python
df['IMPORTANCE'] = df.apply(importance_factor, axis=1)

df.sort_values("IMPORTANCE", ascending=False).head(1)
```

```python
# Organise columns
df["TEAM LIST"] = np.nan

cols = df.columns.tolist()
cols = cols[-3:] + cols[:-3]
df = df[cols]
df.head(5)
```

```python
team_cols = df.loc[:, 'TEAM_01':].columns.tolist()
team_cols_len = len(team_cols)

def create_team(x):
    y = []
    for i in team_cols:
        y.append(x[i])
    return y

df["TEAM LIST"] = df.apply(create_team, axis=1)

df.head(2)
```

```python
team_list = df["TEAM LIST"].tolist()

def find_names(name):
    j = []

    for i in team_list:
        j.append(name in i)
    return j

def name_frame(name):
    p = find_names(name)
    q = df[p]
    return q
```

```python
# Load name list
name_file = 'key-people-list.txt'
file_open = open(name_file)
c = file_open.readlines()
name_list = []
for i in c:
    y = i.strip()
    name_list.append(y)
    
print name_list
```

```python
def process_names(d):
    e = d.split('_')
    f = name_frame(e[1])
    h = len(f["YEAR"].tolist())
    f["BRANCH KEY"] = e[1]
    f["KEY LIST"] = e[0]
    f["KEY TOTAL"] = h
    print "Done: " + d
    return f

count = 0
index = []
columns = []
total_df = pd.DataFrame(index=index, columns=columns)

for g in name_list:
    total_df = total_df.append(process_names(g))
    count += 1
    
save_path = "people_output/_total-people.csv"
total_df.to_csv(save_path, index=False)
print "Processed: " + str(count)
```
