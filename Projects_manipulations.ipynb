```python
# Imports
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import csv
```

```python
# Import data
filename = 'OMA-project-data.csv'
df = pd.read_csv(filename, dtype={'IMG_': str})

#add decade col
def create_decade(x):
    d = str(x['YEAR'])
    e = d[:3] + "0s"
    return e

df['DECADE'] = df.apply(create_decade, axis=1)

#Update image column with paths to B/W photos
bw_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Black and White Images\\"

def create_bw_path(x):
    return bw_path + str(x['IMG_']) + ".jpg"
    
df['@IMG'] = df.apply(create_bw_path, axis=1)

#Remove slashes from Typology Col
def type_clean (x):
    if "/" in str(x["TYPOLOGY_01"]):
        y = str(x["TYPOLOGY_01"]).replace("/", "-")
        return y
    else:
        return x["TYPOLOGY_01"]
        

df["TYPOLOGY_01"] = df.apply(type_clean, axis=1)


#Update image column with paths to Typology backgrounds
type_path = "D:\\Google Drive\\Architecture\\2017\\11222 Critique\\11222 OMA Map\\Project Data\\Typology Backgrounds\\"

def create_type_path(x):
    return type_path + str(x['TYPOLOGY_01']) + ".png"
    
df['@TYPE_BG'] = df.apply(create_type_path, axis=1)

#Drop
df = df.drop(["LINK", "TYPOLOGY_02", "URL STRING", "Unnamed: 11", "IMG_", "@Photos"], axis=1)

# Organise columns
df["TEAM LIST"] = np.nan

cols = df.columns.tolist()
cols = cols[-4:] + cols[:-4]
df = df[cols]

df.head(10)
```

```python
save_path = "total_output/projects.csv"
df.to_csv(save_path, index=False, encoding='utf-16')
print "Saved: " + save_path
```
